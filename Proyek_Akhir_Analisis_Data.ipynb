## RFM Analysis
Analisis lanjutan ini dilakukan dengan tujuan mengetahui lebih lanjut pengguna casual dan registered dengan memperhatikan tiga faktor, juga untuk menambah latihan diluar dua pertanyaan bisnis yang sudah disiapkan.
- **Recency**
  Melihat kapan terakhir pengguna casual dan pengguna registered menggunakan atau melakukan transaksi
- **Frequency**
 Melihat seberapa sering pengguna casual dan pengguna registered melakukan transkasi
- **Monetary**
Melihat seberapa besar revenue yang diberikan dari hasil hasil transaksi pengguna casual dan pengguna registered

Kemudian hasil dari analisis akan digunakan untuk membagi pengguna berdasarkan dengan hasil dari RFM_score. Menampilkan hasil analisis dalam bentuk Visualisasi untuk membantu komunikasi.

```python
rfm_df = all_df.groupby(by="instant", as_index=False).agg({
    "dteday_x": "max", #mengambil tanggal order terakhir
    "casual_y": "nunique",
    "registered_y": "nunique",
    "cnt_y": "sum"
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency_casual", "frequency_registered", "monetary"]
rfm_df.head()
```

```python
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = pd.to_datetime(rfm_df["max_order_timestamp"]).dt.date
recent_date = pd.to_datetime(all_df["dteday_x"]).dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.describe()
```

```python
rfm_df.sort_values(by="recency", ascending=True).head(5)
```

```python
rfm_df.sort_values(by="frequency_casual", ascending=False).head(5)
```

```python
rfm_df.sort_values(by="frequency_registered", ascending=False).head(5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sns.barplot(y="frequency_casual", x="customer_id", data=rfm_df.sort_values(by="frequency_casual", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(y="frequency_registered", x="customer_id", data=rfm_df.sort_values(by="frequency_registered", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Perilaku Pengguna Casual dan Registered Berdasarkan Hasil Analisis RFM  (customer_id)", fontsize=20)
plt.show()
```

### Mengurutkan pengguna berdasarkan recency, frequency, & monetary score

```python
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_casual_rank'] = rfm_df['frequency_casual'].rank(ascending=True)
rfm_df['f_registered_rank'] = rfm_df['frequency_registered'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

rfm_df.head()
```

```python
# normalizing the rank of the customers
rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_casual_rank_norm'] = (rfm_df['f_casual_rank']/rfm_df['f_casual_rank'].max())*100
rfm_df['f_registered_rank_norm'] = (rfm_df['f_registered_rank']/rfm_df['f_registered_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100

rfm_df.drop(columns=['r_rank', 'f_casual_rank','f_registered_rank', 'm_rank'], inplace=True)

rfm_df.head()
```

```python
rfm_df['RFM_score'] = 0.15*rfm_df['r_rank_norm']+0.28 * \
    rfm_df['f_casual_rank_norm']+0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['customer_id', 'RFM_score']].head(7)
```

```python
rfm_df['RFM_score'] = 0.15*rfm_df['r_rank_norm']+0.28 * \
    rfm_df['f_registered_rank_norm']+0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['customer_id', 'RFM_score']].head(7)
```

### Segmentasi Pengguna Berdasarkan RFM_score

```python
rfm_df["customer_segment"] = np.where(
    rfm_df['RFM_score'] > 4.5, "Top customers", (np.where(
        rfm_df['RFM_score'] > 4, "High value customer",(np.where(
            rfm_df['RFM_score'] > 3, "Medium value customer", np.where(
                rfm_df['RFM_score'] > 1.6, 'Low value customers', 'lost customers'))))))

rfm_df[['customer_id', 'RFM_score', 'customer_segment']].head(20)
```

```python
customer_segment_df = rfm_df.groupby(by="customer_segment", as_index=False).customer_id.nunique()
customer_segment_df
```

```python
customer_segment_df['customer_segment'] = pd.Categorical(customer_segment_df['customer_segment'], [
    "lost customers", "Low value customers", "Medium value customer",
    "High value customer", "Top customers"
])
```

```python
plt.figure(figsize=(10, 5))
colors_ = ["#D3D3D3", "#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(
    x="customer_id",
    y="customer_segment",
    data=customer_segment_df.sort_values(by="customer_segment", ascending=False),
    palette=colors_
)
plt.title("Number of Customer for Each Segment", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```
